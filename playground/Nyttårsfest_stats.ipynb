```python
import re
from GENF import GENF
import pandas as pd
from pathlib import Path
pd.set_option('display.max_columns', None)
```

## NyttÃ¥rsfest Gen-F ##

### Setup ###

```python
genf = GENF()
```

### PÃ¥melding ###

```python
path_pÃ¥minnelser = genf._pameldingsmappe
all = pd.DataFrame()
for file in path_pÃ¥minnelser.rglob('*.csv'):
    # print(f'Reading {file.stem}')
    df = pd.read_csv(file)
    pattern = r'(\d{4}.\d{2}.\d{2})'
    event_date = re.search(pattern,file.stem).group(1)
    event_type = file.stem.split('_')[1]
    df['event_date'] = event_date
    df['event_date'] = pd.to_datetime(df['event_date'])
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    df['timestamp_date'] = df['timestamp'].dt.date
    df['timestamp_day'] = df['timestamp'].dt.day
    df['timestamp_dayofweek'] = df['timestamp'].dt.dayofweek
    df['timestamp_hour'] = df['timestamp'].dt.hour
    df['timestamp_minute'] = df['timestamp'].dt.minute
    df['time_to_event'] = df['event_date'] - df['timestamp']
    df['event_type'] = event_type
    all = pd.concat([all,df])
    all['name'] = all['first_name'] + ' ' + all['last_name']
all['answer_bool'] = all['answer'].apply(lambda x: 0 if any(word in str(x).lower() for word in ['ikke', 'nei', 'se']) or 'ğŸ‘€' in str(x) else 1)    
```
